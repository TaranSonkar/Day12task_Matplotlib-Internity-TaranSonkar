**Matplotlib Style**

Matplotlib is the most popular package or library in Python which is used for data visualization. By using this library we can generate plots and figures, and can easily create raster and vector files without using any other GUIs. With matplotlib, we can style the plots like, an HTML webpage is styled by using CSS styles. We just need to import style package of matplotlib library. 

There are various built-in styles in style package, and we can also write customized style files and, then, to use those styles all you need to import them and apply on the graphs and plots. In this way, we need not write various lines of code for each plot individually again and again i.e. the code is reusable whenever required. 

First, we will import the module:

from matplotlib import style

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
from pandas import DataFrame
```

**Solarize_Light2**

```python
x=np.arange(10)
y=x*2
plt.style.use('Solarize_Light2')
plt.plot(x,'o-',y,'-o')
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('Solarize_light2')
plt.show()
```

**'_classic_test_patch'**

```python
x=np.arange(10)
y=x*2
plt.style.use('_classic_test_patch')
plt.bar(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('_classic_test_patch')
plt.show()
```

**bmh**

```python
x=np.arange(10)
y=x*2
plt.style.use('bmh')
plt.scatter(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('bmh')
plt.show()
```

**classic**

```python
x=np.arange(10)
y=np.sin(x)
plt.style.use('classic')
plt.plot(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('classic')
plt.show()
```

**'dark_background'**

```python
x=np.arange(10)
y=np.sin(x)
plt.style.use('dark_background')
plt.plot(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('Dark background')
plt.show()
```

**'seaborn-poster'**

```python
x=np.arange(10)
y=np.sin(x)
plt.style.use('seaborn-poster')
plt.plot(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('seaborn-poster')
plt.show()
```

**'tableau-colorblind10'**

```python
x=np.arange(10)
y=np.cos(x)
plt.style.use('tableau-colorblind10')
plt.plot(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('tableau-colorblind10')
plt.show()
```

**'ggplot'**

```python
x=np.arange(10)
y=np.cos(x)
plt.style.use('ggplot')
plt.plot(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('ggplot')
plt.show()
```

**'grayscale'**

```python
x=np.arange(10)
y=np.cos(x)
plt.style.use('grayscale')
plt.plot(x,y)
plt.xlabel('X-axis',color='red')
plt.ylabel('Y-axis',color='m')
plt.title('grayscale')
plt.show()
```

**SubPlots**

```python
fig,axes=plt.subplots(1,3,figsize=(15,4))
axes
```

```python
x=np.arange(2,50)
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,4))
axes[0].plot(x,np.sin(x),color='m')
axes[0].set_title('Sin')
axes[1].plot(x,np.cos(x),color='y')
axes[1].set_title('cos')
axes[2].plot(x,np.tan(x),color='c')
axes[2].set_title('tan')
plt.show()
```

**Plot multiple charts, subplots and figures**

```python
fig,axes=plt.subplots(1,4,figsize=(18,4))
x=np.arange(10)
plt.style.use('dark_background')
axes[0].plot(x,np.sin(x),color='y')
axes[0].set_title('sin',color='red')

axes[1].plot(x,np.cos(x),color='m')
axes[1].set_title('cos',color='red')

axes[2].plot(x,np.tan(x),color='c')
axes[2].set_title('tan',color='red')

axes[3].plot(x,np.abs(x),color='c') #return positive value
axes[3].set_title('abs',color='red')

plt.show()
```
